```python
datar = data.drop(columns=['요일','구분','지점명'],axis=1)

streets = datar["지점번호"].tolist()
streets = set(streets)

dataGroupedByStreet = []                        
for street in streets:
    dataGroupedByStreet.append(datar[datar["지점번호"]==street]) # 지점명이 같은 row들을 묶어서 JanList라는 array에 집어넣음
    
for street in dataGroupedByStreet:
    count = 0
    streetName = street.iloc[0, 2]
    for index, row in street.iterrows():
        print(index)
        try:
            street.loc[index, "시간"] = row["시간"].split(' ')[1].replace("시","")

            # 500개마다 카운트 할 것
            count += 1
            if count%500 == 0: 
                print(count)
        except:
            print(f"ERROR on {index}")
    
    print(f"{streetName} 끝")

    # int로 바꾸고 sorting
    street = street.astype({"시간":"int8"})
    street.sort_values(by=['일자','시간'], inplace=True)
    street.to_csv(f"{streetName}.csv")
```
