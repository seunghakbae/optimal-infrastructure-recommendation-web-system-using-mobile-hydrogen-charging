```python
import pandas as pd
import numpy as np
from selenium import webdriver
import re
import time
```

```python
jijeom_list= []
```

```python
for i in range(len(jijeom)):
    jijeom_list.append(jijeom.iloc[i]["주소"])
```

```python
jijeom_list[0]
```

```python
driver =webdriver.Chrome()
driver.get("https://map.google.com")
```

```python
jijeom_location = list()
# 정규식으로 처리
p=re.compile(r'(@+.+,)')
i=0

for i in range(len(jijeom)):
    # 주소 뽑는 매크로
    driver.find_element_by_css_selector("#gs_lc50 .tactile-searchbox-input").send_keys(jijeom_list[i])
    time.sleep(0.1)
    driver.find_element_by_class_name("searchbox-searchbutton").click()
    time.sleep(3)
    driver.find_element_by_css_selector("#gs_lc50 .tactile-searchbox-input").clear()
    time.sleep(0.1)
    url=driver.current_url
    time.sleep(0.1)

    #좌표 정리
    location = p.findall(url)
    location=location[0].replace("@", '')
    location=location.split(",")
    del(location[2])
    
    
    jijeom_location.append([[jijeom_list[i]],location])
```

```python
jijeom_raw=jijeom_location
```

```python
jijeom_location[0][0][0]
```

```python
a = []
for i in range(len(jijeom_location)):
    a.append([jijeom_location[i][0][0], jijeom_location[i][1][0],jijeom_location[i][1][1]])
```

```python
jijeom_location=pd.DataFrame(a)
```

```python
jijeom_location.rename(columns={0:"주소", 1:"x", 2:"y"},inplace=True)
```

```python
jijeom=pd.merge(jijeom,jijeom_location,left_index=True, right_index=True)
# taxi_merge = pd.merge(tt,taxi, left_index=True, right_index=True)
```
