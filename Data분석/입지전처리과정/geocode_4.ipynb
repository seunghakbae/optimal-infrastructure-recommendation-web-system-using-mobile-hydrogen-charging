```python
import pandas as pd
import folium
import random
import urllib
import json
import requests
from bs4 import BeautifulSoup
import datetime as dt

missing_data = pd.DataFrame(columns=['Uses', 'address'])
coor_data = pd.DataFrame(columns=['Uses', 'address', 'latitude', 'longitude'])

def get_coor(csv_path, api_key):
    
    big_building = pd.read_csv(csv_path, encoding = "cp949", engine='python')
    
    missing_index = 0
    coor_index = 0
    count = 0
    
    start = dt.datetime.now()
    
    for i in range(len(big_building)):
        
        
        delta= dt.datetime.now()-start
        
        if delta.seconds % 30 == 0 and delta.seconds != 0:
            print("every 30 seconds count : " + str(count))
        
        url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+ big_building.iloc[i]['address']
        headers = {'Authorization': 'KakaoAK {}'.format(api_key)}
        result = json.loads(str(requests.get(url,headers=headers).text))
        
        location = big_building.iloc[i]['A30']
        address =  big_building.iloc[i]['address']
        
        try:
            if result['meta']['total_count'] == 0:
                print(location + " " + address + ' coordinate not found')
                missing_data.loc[missing_index] = [location, address]
                missing_index = missing_index + 1
            else:
                coor_data.loc[coor_index] = [location, address,result['documents'][0]['address']['y'], result['documents'][0]['address']['x']]
                coor_index = coor_index + 1
                
        except:
            print("eror : " + location + " " + address)
            missing_data.loc[missing_index] = [location, address]
            missing_index = missing_index + 1
            
        count += 1
    
    coor_data.to_csv("found_big_building_coor", mode="w", encoding = 'ms949')
    missing_data.to_csv("missing_big_building_coor", mode="w", encoding = 'ms949')       
```
