```python
p = re.compile("((.+?)+[^가-힣]+[0-9])")
```

```python
jucha["도로명"]=list(range(len(jucha)))
```

```python
for i in range(len(jucha)):
    try:
        jucha["도로명"][i]=p.findall(jucha[:]["소재지도로명주소"][i])[0][0]
    except:
        jucha["도로명"][i]='None'
```

```python
jucha["도로명"]
```

```python
for i in range(len(jucha)):
    # 주소 뽑는 매크로
    
    driver.find_element_by_id("ctl00_cphContents_searchTextField").send_keys(str(jucha.도로명[i]))
    time.sleep(0.1)
    driver.find_element_by_id("ctl00_cphContents_ImageButton1").click()
    time.sleep(1)
    try:
        jucha.행정구역[i]=driver.find_element_by_css_selector(".result_data .tbl_type01 tbody tr td:nth-last-child(3)").text
        driver.find_element_by_id("ctl00_cphContents_searchTextField").clear()
        time.sleep(0.1)
    except:
        jucha.행정구역[i]=None
        driver.find_element_by_id("ctl00_cphContents_searchTextField").clear()
        continue
```
