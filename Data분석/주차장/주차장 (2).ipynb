```python
import pandas as pd
from selenium import webdriver
import re
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("http://juso.seoul.go.kr/service/totalsearch/total_search.aspx")
```

```python
mer["행정구역"]=list(range(len(mer)))
```

```python
mer
```

```python
for i in range(756,768):
    # 주소 뽑는 매크로
    
    driver.find_element_by_id("ctl00_cphContents_searchTextField").send_keys(str(mer.도로명주소[i]))
    time.sleep(0.1)
    driver.find_element_by_id("ctl00_cphContents_ImageButton1").click()
    time.sleep(1)
    try:
        mer.행정구역[i]=driver.find_element_by_css_selector(".result_data .tbl_type01 tbody tr td:nth-last-child(3)").text
        driver.find_element_by_id("ctl00_cphContents_searchTextField").clear()
        time.sleep(0.1)
    except:
        mer.행정구역[i]=None
        driver.find_element_by_id("ctl00_cphContents_searchTextField").clear()
        continue
```
