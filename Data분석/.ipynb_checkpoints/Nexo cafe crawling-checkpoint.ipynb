### 10page씩 반복 저장 

150명씩 저장 

```python
for x in range(2,12):
    time.sleep(1)
    driver.find_elements_by_css_selector("#content-area .prev-next a")[x].click()
    time.sleep(0.3)
    temp=[_.text for _ in driver.find_elements_by_css_selector("#content-area .article-board tr > .td_name")]
    for x in temp:
        nexo_list.append(x)  
```

```python
while True:
    if driver.find_elements_by_css_selector("#content-area .prev-next a")==False:
        break
    for x in range(2,12):
        time.sleep(0.4)
        driver.find_elements_by_css_selector("#content-area .prev-next a")[x].click()
        time.sleep(0.3)
        temp=[_.text for _ in driver.find_elements_by_css_selector("#content-area .article-board tr > .td_name")]
        for x in temp:
            nexo_list.append(x) 
```

```python
import sqlite3
conn = sqlite3.connect('nexo.db')
cur = conn.cursor()
```

```python
cur.execute('''
    CREATE TABLE region(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        region TEXT,
        count INTEGER
    );
''')
```

```python
filter_region=[]
```

```python
for x in final_nexo_list:
    if ('N' in x):
        s = x.split('N')
        filter_region.append(s[0])
```

```python
len(filter_region)
```

```python
len(set(filter_region))
```

```python
final_nexo = []
```

```python
for x in set(filter_region):
    cnt=filter_region.count(x)
    final_nexo.append((x,cnt))
```

```python
final_nexo
```

```python
cur.executemany('''
    INSERT INTO region(region, count) VALUES(?,?);
''',final_nexo)
```
