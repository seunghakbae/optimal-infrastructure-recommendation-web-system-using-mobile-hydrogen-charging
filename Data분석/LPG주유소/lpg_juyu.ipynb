```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
lpg = pd.read_csv("lpg_juyuso.csv", encoding='cp949')
```

```python
lpg["crawl"]=list(range(0,len(lpg)))
```

```python
lpg["x"]=list(range(len(lpg)))
```

```python
lpg["y"]=list(range(len(lpg)))
```

```python
lpg.crawl[1]
```

```python
for i in range(len(lpg)):
    req=p.findall(lpg[:]["주소"][i])[0][0]
    lpg.crawl[i]=req
```

```python
lpg.drop(columns=["지역", "관리구분","전화번호"],inplace =True)
```

```python
driver.get("https://map.google.com")
```

```python
k=re.compile(r'(@+.+,)')
for i in range(len(lpg)):
    
    driver.find_element_by_css_selector("#gs_lc50 .tactile-searchbox-input").send_keys(str(lpg.crawl[i]))
    time.sleep(0.1)
    driver.find_element_by_class_name("searchbox-searchbutton").click()
    time.sleep(2.5)
    driver.find_element_by_css_selector("#gs_lc50 .tactile-searchbox-input").clear()
    time.sleep(0.1)
    url=driver.current_url
    time.sleep(0.1)

    #좌표 정리
    location = k.findall(url)
    location=location[0].replace("@", '')
    location=location.split(",")
    
    lpg.x[i]=location[0]
    lpg.y[i]=location[1]
    
    print(lpg.x[i], lpg.y[i])
```

```python
lpg.to_csv("lpg_location.csv",encoding="cp949")
```
