```python
location_list=list()
# 좌표 데이터 넣기

for i in range(36111,36444):
    # 좌표 검색하기
    driver.find_element_by_id("search.keyword.query").send_keys(str(taxi.iloc[i][4])+","+str(taxi.iloc[i][3]))
    time.sleep(0.1)
    driver.find_element_by_id("search.keyword.submit").click()
    time.sleep(0.1)
    location=driver.find_element_by_css_selector("#localInfo").text.split()
    #새로고침 or 써져있는 거 지우기
    time.sleep(0.1)
    driver.find_element_by_id("search.keyword.query").clear()
    #좌표 추가
    location.append(taxi.iloc[i][0])
    location.append(taxi.iloc[i][1])
    location.append(taxi.iloc[i][2])
    #동 이름 list추가 
    location_list.append(location)
```

```python
pd.DataFrame(location_list).to_csv("h.csv", encoding = "cp949")
```
