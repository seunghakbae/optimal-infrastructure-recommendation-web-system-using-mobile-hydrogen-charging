```python
import pandas as pd
import random
import json
import folium
```

```python
# 동 랜덤정보
dong = pd.read_csv("행정동.csv")
dong.rename(columns={'면적':'데이터'}, inplace=True)
dong.iloc[7,2]="종로1·2·3·4가동"
for index, row in dong.iterrows():
    dong.iloc[index,3]=random.randint(1, 50)
dong.insert(0, '행정구역코드', None)
dong.drop(columns=["Unnamed: 0"], inplace=True)
dong.iloc[63,2]="금호2·3가동"
dong.iloc[108,2]="면목3·8동"
dong.iloc[172,2]="상계3·4동"
dong.iloc[173,2]="상계6·7동"
dong.iloc[174,2]="중계2·3동"
# dong = dong.astype('str')
```

```python
seoul_map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5, tiles="cartodbpositron")
seoul_map.choropleth(geo_data=geo_str, 
                     data = dong,
                     columns=['행정구역코드', '데이터'],
                     fill_color='PuRd', 
                     key_on='properties.code',
                     highlight=True,
                     fill_opacity=0.5,
                     line_opacity=1,
                     legend_name="myData")

seoul_map.save("seoul_map.html")
```
