```python
import pandas as pd
import random
import json
import folium
```

```python
# 동 랜덤정보
data = pd.read_csv("fifi.csv")
data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
dongNumber=pd.read_csv("행정동분류.csv")
dongNumber.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
dongNumber.행정구역명=dongNumber.행정구역명.astype('str')  
```

```python
for index, row in data.iterrows():
    for index2, row2 in dongNumber.iterrows():
        if data.loc[index,"동"] == dongNumber.loc[index2,"행정구역명"]:
            data.loc[index,"행정구역코드"] = dongNumber.loc[index2,"행정구역코드"]
            break
print("done")
```

```python
seoul_map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5, tiles="cartodbpositron")
seoul_map.choropleth(geo_data=geo_str, 
                     data = data ,
                     columns=['동', '백화점_개소'],
                     fill_color='RdPu',  # GnBu
                     key_on='properties.name',
                     highlight=True,
                     fill_opacity=0.5,
                     line_opacity=1,
                     legend_name="백화점")
seoul_map.choropleth(geo_data=geo_str, 
                     data = data ,
                     columns=['동', '합계_사업체수'],
                     fill_color='RdPu',  # GnBu
                     key_on='properties.name',
                     highlight=True,
                     fill_opacity=0.5,
                     line_opacity=1,
                     legend_name="합계_사업체수")
seoul_map.choropleth(geo_data=geo_str, 
                     data = data ,
                     columns=['동', '㎡당 매매평균가'],
                     fill_color='RdPu',  # GnBu
                     key_on='properties.name',
                     highlight=True,
                     fill_opacity=0.5,
                     line_opacity=1,
                     legend_name="㎡당 매매평균가")
seoul_map.choropleth(geo_data=geo_str, 
                     data = data ,
                     columns=['동', '수입차합계'],
                     fill_color='RdPu',  # GnBu
                     key_on='properties.name',
                     highlight=True,
                     fill_opacity=0.5,
                     line_opacity=1,
                     legend_name="수입차합계")
seoul_map.choropleth(geo_data=geo_str, 
                     data = data ,
                     columns=['동', '인구밀도'],
                     fill_color='RdPu',  # GnBu
                     key_on='properties.name',
                     highlight=True,
                     fill_opacity=0.5,
                     line_opacity=1,
                     legend_name="인구밀도")
seoul_map.save("size.html")
```
