```python
import numpy as np
import pandas as pd
```

# LPG 데이터 프레임 만들기

```python
# DB로 만들 LPG_DATAFRAME만들기. 
lpg_dataframe = pd.DataFrame(columns=['lpg_id', 'name', 'x_cor', 'y_cor', 'installed_2020', 'installed_2022', 'installed_2024', 'installed_2026', 'installed_2028', 'dong_id', '동'])
```

```python
# lpg 전체 데이터를 불러온다. 
lpg = pd.read_csv("raw_Data/lpg.csv", encoding='cp949')
```

```python
# 겹치는 것은 뺀다. 
lpg = lpg.drop_duplicates(subset='업체명')
```

```python
# 이름 데이터를 뺸다. 
lpg_name = lpg['업체명']
```

```python
# 경도 데이터를 뺀다. 
lpg_x_cor = lpg['경도']
```

```python
# 위도 데이터를 뺀다. 
lpg_y_cor = lpg['위도']
```

```python
# 동 데이터를 뺸다. 
lpg_dong = lpg['동']
```

```python
# 자치구 데이터를 뺀다 .
lpg_gu = lpg['자치구']
```

```python
# 이름 데이터 넣기.
lpg_dataframe['name'] = lpg_name
```

```python
# lpg_id 넣기.
lpg_dataframe['lpg_id'] = list(range(0,len(lpg_dataframe)))
```

```python
# x_cor넣기.
lpg_dataframe['x_cor'] = lpg_x_cor
```

```python
# y_cor넣기.
lpg_dataframe['y_cor'] = lpg_y_cor
```

```python
# 동 넣기.
lpg_dataframe['동'] = lpg_dong
```

```python
lpg_dataframe['자치구'] = lpg_gu
```

```python
# 2020, 2022, 2024, 2026, 2028 다 0으로 채우기.
for year in [2020, 2022, 2024, 2026, 2028]:
    lpg_dataframe["installed_" + str(year)] = 0
```
